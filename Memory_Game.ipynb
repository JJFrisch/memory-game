```python
import math
import time
from graphics import *
import random
board = []  #positive numbers will be hidden, negitive numbers will be shown
map = []
winH = 600
winW = 600
phase = "1" # 1-Waiting for 1st click, 2-displaying one click waiting for 2nd click, 3rd checking match



def createBoard():
    global board
    w =  int(input("What width?"))
    h =  int(input("What height?"))
    if w > 1 and h > 1:
        highestNum = int(w*h/2)
        vals = []
        for num in range(highestNum):
            vals.append(num+1)
            vals.append(num+1)
        for y in range(h):
            row = []
            for x in range(w):
                if len(vals) > 0:
                    val = random.choice(vals)
                    vals.remove(val)
                    row.append(val)
            board.append(row)
    else: 
        createBoard()


def allFound():
    for y in range(len(board)):
        for x in range(len(board[y])):
            if board[y][x] > 0:
                return False
    return True

def clickPos(p):
    rw = winW/len(board[0])
    rh = winH/len(board)
    col = math.floor(p.getX()/rw)
    row = math.floor(p.getY()/rh)
    return [col,row]


# display the win screen
def winDisplay(win, turns):
    for item in win.items: # undraw all old items
            item.undraw()
    r = Rectangle(Point(0,0), Point(winW,winH))
    r.setFill("light green")
    t = Text(Point(winW/2,winH/2-90), "You won, all pairs were found in " + str(turns)+ " turns!")
    t2 = Text(Point(winW/2,winH/2), "Good Job!")
    t.setSize(20)
    t2.setSize(30)
    r.draw(win)
    t2.draw(win)
    t.draw(win)

def graphicsMain():
    global phase
    createBoard()
    print(board)
    win = GraphWin("2D Robot", winW, winH, False)
    click1 = None
    click2 = None
    playing = True
    turns = 0
    while playing:
        update(15)
        display(win)
        if allFound():
            playing = False
        
        if phase == "1":
            mouse = win.getMouse()
            # un comment this code and recoment out phase 3 to change the way it works
            # if click1 != None and click2 != None and board[click1[1]][click1[0]] != board[click2[1]][click2[0]]:
            #     board[click1[1]][click1[0]] = board[click1[1]][click1[0]] * -1
            #     board[click2[1]][click2[0]] = board[click2[1]][click2[0]] *- 1
            click1 = clickPos(mouse)
            if board[click1[1]][click1[0]] > 0:
                board[click1[1]][click1[0]] = board[click1[1]][click1[0]] * -1
                phase = "2"
        elif phase == "2":
            mouse = win.getMouse()
            click2 = clickPos(mouse)
            if board[click2[1]][click2[0]] > 0 and click1 != click2:
                board[click2[1]][click2[0]] = board[click2[1]][click2[0]] * -1
                phase = "3"
        elif phase == "3":
            if board[click1[1]][click1[0]] != board[click2[1]][click2[0]]:
                mouse = win.getMouse()
                board[click1[1]][click1[0]] = board[click1[1]][click1[0]] * -1
                board[click2[1]][click2[0]] = board[click2[1]][click2[0]] *- 1
            phase = "1"
            turns += 1

    winDisplay(win, turns)
    win.getMouse()
    win.close()

        
def display(win):
    global map
    for item in win.items: # undraw all old items
        item.undraw()
    rw = winW/len(board[0])
    rh = winH/len(board)
    for row in range(len(board)):
        for col in range(len(board[row])):
            rx = col*rw
            ry = row*rh
            r = Rectangle(Point(rx,ry), Point(rx+rw,ry+rh))
            if board[row][col] < 0:
                r.setFill("white")
            else:
                r.setFill("light grey")
            r.setOutline("black")
            r.draw(win)
            
            if board[row][col] < 0:
                t = Text(Point(rx+rw/2,ry+rh/2), abs(board[row][col]))
                t.setFill("black")
                t.setSize(20)
                t.draw(win)
                
if __name__ == '__main__':
    graphicsMain()
```
